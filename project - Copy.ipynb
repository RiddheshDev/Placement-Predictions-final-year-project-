```python
import scipy.stats as stats
chi2_results = pd.DataFrame(columns=['Variable', 'Chi-square', 'Degrees of Freedom', 'p-value'])

# Iterate over each column in the DataFrame (assuming all columns are categorical variables)
for column in x_categories_df.columns:
    # Create a contingency table of observed frequencies
    contingency_table = pd.crosstab(x_categories_df[column], y)

    # Perform the chi-square test for independence
    chi2, p_value, dof, _ = stats.chi2_contingency(contingency_table)

    # Append the results to the chi2_results DataFrame
    chi2_results = chi2_results.append({'Variable': column,
                                        'Chi-square': chi2,
                                        'Degrees of Freedom': dof,
                                        'p-value': p_value}, ignore_index=True)

# Print the chi-square results
print(chi2_results)
```

# ROCAUC curve

```python
from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt

# Fit and predict probabilities for each model
models = [log_reg, forest, tree, xgb,knn,clf]
model_names = ["Logistic Regression", "Random Forest", "Decision Tree", "XGBoost",'KNN',"SVM"]
colors = ["b", "g", "r", "c",'y','k']  # Define colors for each model

plt.figure(figsize=(8, 6))  # Set the figure size

# Plot ROC curve for each model
for model, name, color in zip(models, model_names, colors):
    y_test_pred = model.predict(x_test)  # Predict probabilities of positive class
    fpr, tpr, thresholds = roc_curve(y_test,y_test_pred )
    auc = roc_auc_score(y_test, y_test_pred)
    plt.plot(fpr, tpr, color=color, label=f"{name} (AUC = {auc:.2f})")

# Plot the random guess line
plt.plot([0, 1], [0, 1], "k--", label="Random Guess")

# Set labels and title
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Receiver Operating Characteristic (ROC) Curve")

# Set legend and grid
plt.legend(loc="lower right")
plt.grid(True)

# Show the plot
plt.show()
```
